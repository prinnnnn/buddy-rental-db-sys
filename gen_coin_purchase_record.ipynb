```python
import datetime
import random
import json
import pandas as pd
from faker import Faker
import faker as f
```

```python
def save_to_csv(data, table_name, base_filename="mock_data"):
    pd.DataFrame(data).to_csv(f"./data/{base_filename}_{table_name}.csv", index=False, header=False)
    print(f"Data saved to ./data/{base_filename}_{table_name}.csv")
```

```python
customer_df = pd.read_csv("./data/mock_data_customer.csv").values
customer_ids = list(customer_df[:, 0])

fake = Faker()
```

```python
payment_methods = ('MobileBanking','Credit/DebitCard','QRPayment')

def generate_payment_method():
    return random.choice(payment_methods)

coin_packages = [100,200,300,500,800,1000,1500,2000,3500]

def generate_coin_amount():
    return random.choice(coin_packages)
```

```python
def generate_mock_coin_purchase_record():

    customer_id = random.choice(customer_ids)
    coin_amount = generate_coin_amount()

    coin_purchase_record = {
        'purchase_id': fake.uuid4(),
        'payment_method': generate_payment_method(),
        'timestamp': datetime.datetime.now().__str__(),
        'cash_price': coin_amount,
        'coin_amount': coin_amount,
        'customer_id': customer_id,
    }

    return coin_purchase_record

def generate_batches_coin_purchase_record(num):
    return [generate_mock_coin_purchase_record() for _ in range(num)]
```

```python
mock_coin_purchase_records = generate_batches_coin_purchase_record(5000)

df = pd.DataFrame(mock_coin_purchase_records)
```
