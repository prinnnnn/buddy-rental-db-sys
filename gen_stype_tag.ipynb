```python
def save_to_csv(data, table_name, base_filename="mock_data"):
    pd.DataFrame(data).to_csv(f"./data/{base_filename}_{table_name}.csv", index=False, header=False)
    print(f"Data saved to ./data/{base_filename}_{table_name}.csv")
```

```python
customer_df = pd.read_csv("./data/mock_data_customer.csv").values
customer_ids = list(customer_df[:, 0])

buddy_df = pd.read_csv("./data/mock_data_buddy.csv").values
buddy_ids = list(buddy_df[:, 0])
```

```python
service_types = [
    "Sports",
    "Video games",
    "Board games",
    "Movie",
    "Hiking",
    "Volunteering",
    "Field trip",
    "Scooba diving",
    "Swimming",
    "Gardening",
    "Fishing",
    "Concerts",
    "Cooking",
    "Choirs"
]

print(len(service_types))
```

```python
service_types = { 
    "type_id": [i+1 for i in range(len(service_types))],
    "description": service_types
}
```

```python
def rand_stype(n,p):
    return stats.binom.rvs(n,p)
```

```python
n = 14
p = 6/14
print(rand_stype(n,p))
```

```python
prefer_service_type = list()
n = 14
p = 3/14
for cid in customer_ids:
    stype_amount = rand_stype(n,p)
    used_stype = set()
    for i in range(stype_amount):
        stype = random.randint(1,14)
        if(stype not in used_stype):
            row = [cid,stype]
            prefer_service_type.append(row)
            used_stype.add(stype)

df = pd.DataFrame(prefer_service_type)
```

```python
df2 = pd.DataFrame.sort_values(df,[0,1])
```

```python
save_to_csv(df, "prefer_service_type")
```

```python
tags = ["gaming", "chilling", "family", "friendly", "travel", "talkative",
       'cute','fighter','arcade','married','uncle','brother','sister',
       'advisor','listener','tutor','furry','helicopter','singer']
print(len(tags))
```

```python
def rand_tag(n,p):
    return stats.binom.rvs(n,p)
```

```python
n = 19
p = 5/19
print(rand_tag(n,p))
```

```python
customer_preferred_tag = list()
n = 19
p = 5/19
for cid in customer_ids:
    tag_amount = rand_tag(n,p)
    used_tag = set()
    for i in range(tag_amount):
        idx = random.randint(0,18)
        tag = tags[idx]
        if(tag not in used_tag):
            row = [cid,tag]
            customer_preferred_tag.append(row)
            used_tag.add(tag)

df = pd.DataFrame(customer_preferred_tag)
```

```python
df
```

```python
df2 = pd.DataFrame.sort_values(df,[0,1])
df2[:15]
```

```python
save_to_csv(df, "customer_preferred_tag")
```

```python
buddy_preferred_tag = list()
n = 19
p = 9/19
for bid in buddy_ids:
    tag_amount = rand_tag(n,p)
    used_tag = set()
    for i in range(tag_amount):
        idx = random.randint(0,18)
        tag = tags[idx]
        if(tag not in used_tag):
            row = [bid,tag]
            buddy_preferred_tag.append(row)
            used_tag.add(tag)

df = pd.DataFrame(buddy_preferred_tag)
```

```python
df2 = pd.DataFrame.sort_values(df,[0,1])
df2[:15]
```
